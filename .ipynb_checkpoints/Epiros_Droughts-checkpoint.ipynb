```python
#more Matplotlib setups

plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
})
```

```python
df = pd.read_csv('hpeiros.csv')
```

```python
# Filtering only the month of October
oct_df=df[df.DATA.str.contains( 'Oct',na=False)]
```

```python
# Keeping the columns with the Date and the index I want
oct_df=oct_df[['DATA','SPEI_12']]
```

```python
# Keeping the year frome the date
oct_df['date'] = pd.to_datetime(oct_df['DATA'])
```

```python
oct_df['year'] = oct_df['date'].dt.year
```

```python
oct_df=oct_df.drop(columns=['date','DATA'])
```

```python
oct_df=oct_df.reset_index().drop(['index'],axis=1)
```

```python
# Making the graph 

sp12 = oct_df.plot(kind='bar',x='year',y='SPEI_12',figsize=(20,5),rot=45, color='yellowgreen', alpha=.8, legend=True)

# putting a defining line in -1 which shows the value from which droughts start
y_value = -1  
sp12.axhline(y=y_value, color='dimgray', linestyle='--', label=f'Y = {y_value}')


sp12.set_title('Epirus Drought Events 1950-2023', fontsize=15, fontweight='bold')
#set a subtitle
plt.suptitle('12-month average of SPEI', fontsize=11, color='dimgray', fontweight='bold')
```

```python
p12 = oct_df.plot(x='year', y='SPEI_12',figsize=(10,5),rot=45, color='yellowgreen', alpha=.8, legend=True)
p12.set_title('A single plot')

y_value = -1  
p12.axhline(y=y_value, color='dimgray', linestyle='--', label=f'Y = {y_value}')
```

```python
p12 = oct_df.plot(x='year',y='SPEI_12',figsize=(10,5),rot=45, color='yellowgreen', alpha=.8, legend=True)

y_value = -1  
p12.axhline(y=-1, color='dimgray', linestyle='--', label=f'Y = {y_value}')
p12.axhline(y=0, color='dimgray',  label=f'Y = {y_value}')


p12.fill_between(oct_df['year'], oct_df['SPEI_12'], y_value, where=(oct_df['SPEI_12'] < y_value),
                 interpolate=True, color='yellowgreen', alpha=0.5)
p12.set_title('Epirus')
p12.legend()

plt.show()
```

```python
oct_df = oct_df.rename(columns={oct_df.columns[1]:'epirus'})
```

```python
oct_df.to_csv('epirus_df.csv',index=False)
```
