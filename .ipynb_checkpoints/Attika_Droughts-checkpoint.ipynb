```python
import pandas as pd
import os
pd.set_option('display.max_rows', 1500)
pd.set_option('display.max_columns', 60)
pd.__version__

import matplotlib.pyplot as plt

#setups
%matplotlib inline
```

```python
def read_csv(filename):
    try:
        df = pd.read_csv(filename, header=1)
    except Exception as e:
        df='chrysoula'
        
        print(f"cannot read {filename}. Error: {e}")
    return df
```

```python
def read_all_csv(path):
    ''' This function:
     - reads all the data of forest fires
     - concatenates them
     - removes unwanted columns
     - turns strings into numbers
     '''
#     this is where we'll save all the dataframes when we read them
    list_of_dfs=[]
#     lists the files of the path
    files = os.listdir(path)
#     filters the files we want to read
    csv_files =[f for f in files if f.endswith('.csv')]
#     loop through the files
    for file in csv_files:
#          prints message for us to see the progress
        print(f"Reading file: {file}")
#          reads the each file
        df=pd.read_csv(f"{path}/{file}")
#         print(df.columns)
#          turn strings into numbers (but keep the categorical strings into strings)
#         df = df.apply(lambda x: pd.to_numeric(x.astype(str), errors='ignore'))
        list_of_dfs.append(df)
        attica = pd.concat(list_of_dfs).reset_index(drop=True)
#        attica = attica[selection] 
    
    return attica
```

```python
df = read_all_csv('C:\\Users\\c.marinou\\Documents\\GitHub\\Purkagies\\attika')
```

```python
df
```

```python
# Filtering only the month of October
oct_df=df[df.DATA.str.contains( 'Oct',na=False)]
```

```python
# Keeping the columns with the Date and the index I want
oct_df=oct_df[['DATA','SPEI_12']]
```

```python
# Keeping the year frome the date
oct_df['date'] = pd.to_datetime(oct_df['DATA'])
```

```python
oct_df['year'] = oct_df['date'].dt.year
```

```python
oct_df=oct_df.drop(columns=['date','DATA'])
```

```python
oct_df=oct_df.reset_index().drop(['index'],axis=1)
```

```python
oct_df=oct_df.groupby('year').mean().reset_index()
oct_df
```

```python
# Making the graph 

sp12 = oct_df.plot(kind='bar',x='year',y='SPEI_12', figsize=(20,5),rot=45, color='pink', alpha=.8, legend=True)

# putting a defining line in -1 which shows the value from which droughts start
y_value = -1  
sp12.axhline(y=y_value, color='dimgray', linestyle='--', label=f'Y = {y_value}')


sp12.set_title('Attica Drought Events 1950-2023', fontsize=15, fontweight='bold')
#set a subtitle
plt.suptitle('12-month average of SPEI', fontsize=11, color='dimgray', fontweight='bold')

plt.savefig('attica_bar.png')
```

```python
p12 = oct_df.plot(x='year',y='SPEI_12',figsize=(10,5),rot=45, color='pink', alpha=.8, legend=True)

y_value = -1  
p12.axhline(y=-1, color='dimgray', linestyle='--', label=f'Y = {y_value}')
p12.axhline(y=0, color='dimgray',  label=f'Y = {y_value}')


p12.fill_between(oct_df['year'], oct_df['SPEI_12'], y_value, where=(oct_df['SPEI_12'] < y_value),
                 interpolate=True, color='pink', alpha=0.5)
p12.set_title('Attica Drought Events 1950-2023', fontweight='bold')
# p12.legend()

plt.savefig('attica_line.png')
```
