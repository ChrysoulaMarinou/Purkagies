```python
file_path = 'C:\\Users\\c.marinou\\Documents\\GitHub\\Purkagies\\krhth.csv'
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
df
```

```python
df['DATA']= df['DATA'].astype(str)
```

```python
df['year'] = df['DATA'].str.extract(r'(\d{4})')
```

```python
df_group = df.drop('DATA',  axis=1)
```

```python
df_group= df_group.groupby('year').median()
```

```python
df_group= df_group.reset_index()
```

```python
df_group_sort=df_group.groupby('year').mean().sort_values('SPEI_48')
df_group_sort
```

```python
df_group_sort = df_group_sort.reset_index()
df_group_sort
```

```python
sp48 = df_group.plot(kind='bar',x='year',y='SPEI_48',figsize=(20,5),rot=45, color='blue', alpha=.8, legend=True)

sp48.set_title('Crete Drought Events 1950-2023', fontsize=15, fontweight='bold')
#set a subtitle
plt.suptitle('48-month median of SPEI', fontsize=11, color='dimgray', fontweight='bold')
```
